To test the effectiveness of these tools, we will conduct small-scale trials at the Universidad Icesi campus, as well collaborate with local professionals and volunteers to conduct fieldwork in two localities at the Parque Natural Nacional Farallones, located on the western Andes, Colombia. One will be located at the lowland wet forest of Alto Anchicaya (700m) one of the rainiest areas in the planet and the other one at a cold humid montane cloud forest located at the University research station, Zygia (2,300m).

```python
import folium
from folium import plugins

# Crear el mapa principal centrado en los puntos y con zoom para mostrar toda Colombia
main_map = folium.Map(location=[4, -72], zoom_start=7)

# Agregar marcadores para cada punto
points = [
    {'name': 'Yatacué- Anchicayá', 'coords': [3.57472, -76.87777]},
    {'name': 'Estación Zygia', 'coords': [3.44050, -76.650]},
    {'name': 'Universidad ICESI', 'coords': [3.34166, -76.52914]}
]
for point in points:
    folium.CircleMarker(
        location=point['coords'],
        radius=5,
        color='red',
        fill=True,
        fill_color='red',
        fill_opacity=1,
        tooltip=f"Coordenadas: {point['coords'][0]}, {point['coords'][1]}"
    ).add_to(main_map)

# Crear el mapa de referencia de Colombia
reference_map = folium.Map(location=[4, -72], zoom_start=6)

# Agregar el mapa de referencia como capa de teselas al mapa principal
folium.TileLayer(tiles='Stamen Terrain', overlay=True).add_to(main_map)
folium.TileLayer(tiles='CartoDB positron', overlay=True).add_to(reference_map)


# Agregar la rosa de los vientos al mapa principal
folium.plugins.FloatImage('https://raw.githubusercontent.com/SECOORA/static_assets/master/maps/img/rose.png', bottom=5, left=5).add_to(main_map)

# Agregar el mapa de referencia como control en la esquina superior derecha del mapa principal
minimap = plugins.MiniMap(toggle_display=True)
main_map.add_child(minimap)

# Mostrar el mapa principal
main_map
```
